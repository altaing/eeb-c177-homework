1. import csv and re
2. load file
3. convert to csv
4. separate authors (by columns)
5. remove parentheses
6. replace " &" with ","
7. remove "'"
8. remove digits
9. split list at ","
10. count occurances

```python
import csv
import re
auth = []
with open('bee_list.txt', 'r') as bee:
    data = csv.DictReader(bee, delimiter='\t')
    header = data.fieldnames
    for row in data:
        auth.append(row['Taxon Author'])
        
    author = str(auth).replace("(","")
    author1 = str(author).replace(")","")
    author2 = str(author1).replace(" &",",")
    author3 = str(author2).replace("'","")
    author4 = re.sub(r'[\s\d]','',author3)
    author5 = str(author4).split(',')
    authorfix = list(filter(None, author5)) #2/27
    
from collections import defaultdict
authorcount = defaultdict(int)
for a in authorfix:
    authorcount[a] += 1
authorcount
#Cockerell has the highest count
```
