```python
import csv
import re
auth = []
with open('bee_list.txt', 'r') as bee:
    data = csv.DictReader(bee, delimiter='\t')
    header = data.fieldnames
    for row in data:
        auth.append(row['Taxon Author'])
auth


    
```

```python
autho = str(auth).replace("(","")
author = str(autho).replace(")","")
author2 = str(author).replace(" &",",")
author3 = re.sub(r'[\s\d]','',author2)
author4 = str(author3).split(',')

from collections import defaultdict
count=defaultdict(int)
for a in author4:
    count[a] +=1
count
```
