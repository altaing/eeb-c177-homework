# 1.1

```python
import numpy
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt(fname = 'inflammation-01.csv', delimiter = ',')
```

```python
np.std(data)
```

```python
pat_std = np.std(data, axis = 0)
plt_std = plt.plot(pat_std)
plt.xlabel('time (days)')
plt.ylabel('standard deviation')
plt.title('standard deviation of inflamtation per day')
plt.show
```

```python
mx_inf = np.max(data, axis=0)
mn_inf = np.min(data, axis=0)
```

```python
plt.plot(mx_inf) #blue
plt.plot(mn_inf) #orange
plt.plot(pat_std) #green
plt.xlabel('time (days)')
```
