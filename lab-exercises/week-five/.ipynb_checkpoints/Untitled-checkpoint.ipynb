```python
import numpy as np
data = np.loadtxt(fname = 'inflammation-01.csv', delimiter = ',')
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
image = plt.imshow(data)
plt.show()
```

```python
avg_inf = np.mean(data, axis=0)
avg_plt = plt.plot(avg_inf)
plt.xlabel('time (days)')
plt.ylabel('mean inflammation score')
plt.title('patient data')
plt.show()
```

```python
mx_inf = np.max(data, axis=0)
mx_plt = plt.plot(mx_inf)
plt.xlabel('time (days)')
plt.ylabel('max inflammation score')
plt.title('patient data')
plt.show()
```

```python
max_plot = plt.plot(np.max(data, axis=0))
plt.show()
```

```python
min_plot = plt.plot(np.min(data, axis=0))
plt.show()
```

1) Create a figure object (width, height)<br>
2) (1 row, 3 columns) (for example)

```python
fig = plt.figure(figsize=(10.0, 3.0))

fig.suptitle('Title', va='bottom', fontsize = 16, color = 'maroon')

subplot1 = fig.add_subplot(1, 3, 1) #rows, cols, position
subplot2 = fig.add_subplot(1, 3, 2)
subplot3 = fig.add_subplot(1, 3 ,3)

subplot1.set_ylabel('average')
subplot1.plot(np.mean(data, axis=0), '--')

subplot2.set_xlabel('time (days)')
subplot2.set_ylabel('max')
subplot2.plot(np.max(data, axis=0), '-o')

subplot3.set_ylabel('min')
subplot3.plot(np.min(data, axis=0), '*')

fig.tight_layout()

plt.show()
```

```python
word = 'lead'
for c in word:
    print(c)

print("\n")    

for c in range(0, len(word)):
    print(word[c])
    
print("\n")

for c in range(0, len(word)):
    print(c)
    
print("\n")
    
for c in range(0, len(word)):
    print(c, word[c])
    
print("\n")
    
for counter, c in enumerate(word):
    print(counter, c)

print("\n")    

for counter, c in enumerate(word):
    print(counter+1, c)
```

# Lists

```python
odds = [1, 3, 5, 7]
print ('odds are:', odds)

print('first element:', odds[0])

print('last element:', odds[3])
print('last element:', odds[-1])

print('second to last element:', odds[-2])
```

```python
for numbers in odds:
    print(numbers)
    
print('\n')
    
for numbers in range(0, len(odds)):
    print(odds[numbers])
```

```python
num = 101
num2 = 101
if num > 100:
    print('greater')
elif num == num2:
    print ('equal')
elif num == 42:
    print('youve solved the universe')
else: print('less')
```

```python
## get all even numbers from 1 - 100
evens = []
odds = []
for ii in range(1,101):
    if (ii % 2 == 0):
        evens.append(ii)
    elif (ii % 2 == 1):
        odds.append(ii)
        
print(odds)
print('\n')
print(evens)
```

```python
evens = [i for i in range(101) if i % 2 == 0]
print(evens)
```

```python
square_nums_10 = [i * i for i in range(20) if i % 2 == 0]
square_nums_10
```
