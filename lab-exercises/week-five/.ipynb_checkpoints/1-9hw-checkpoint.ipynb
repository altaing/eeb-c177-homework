# 1.1

```python
import numpy
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt(fname = 'inflammation-01.csv', delimiter = ',')
```

```python
np.std(data)
```

```python
pat_std = np.std(data, axis = 0)
plt_std = plt.plot(pat_std)
plt.xlabel('time (days)')
plt.ylabel('standard deviation')
plt.title('standard deviation of inflamtation per day')
plt.show
```

# 1.2
### literally placed on top of one another and vetrically stacked

```python
mx_inf = np.max(data, axis=0)
mn_inf = np.min(data, axis=0)

plt.plot(mx_inf) #blue
plt.plot(mn_inf) #orange
plt.plot(pat_std) #green
plt.xlabel('time (days)')
```

```python
stk = plt.figure(figsize = (5, 10))
stk.suptitle('std, max, min', va='bottom', fontsize=16)

sbp1 = stk.add_subplot(3, 1, 1) #rows, cols, pos
sbp2 = stk.add_subplot(3, 1, 2)
sbp3 = stk.add_subplot(3, 1, 3)

sbp1.set_ylabel("std")
sbp1.plot(pat_std, color = "maroon")

sbp2.set_ylabel("max")
sbp2.plot(mx_inf, color = "teal")

sbp3.set_ylabel("min")
sbp3.plot(mn_inf, color = "green")
sbp3.set_xlabel("time (days)")

stk.tight_layout()
plt.show()
```

# 2.2.4

```python
def rev(text):
    txt = ""
    for l in text:
        txt = l + txt
    return txt

print(rev("Ucla"))
```

# 2.2.5

```python
x = 5
coefs = [2, 4, 3]
y = coefs[0] * x**0 + coefs[1] * x**1 + coefs[2] * x**2
print(y)
print('\n')

x = 5
y = 0
coefs = [2, 4, 3]
for idx, val in enumerate(coefs):
    y = y + val * x**idx
    print(y)
```

# 3.1

A list in python is a data set that can be changed(mutable) and ordered. It is denoted by square brackets []

# 3.2

To create a list, you can input values within square brackets [] to which to have assigned to a variable. You can also use the function <code>list()</code>. Separate each item with commas.

# 3.3

To change an element within a list, do a form of the following... <code>list[element]= new_element</code>

# 3.4

One way to append values to an existing list is by using the <code>append()</code> function. For example, <code> mylist.append(new_element)</code>

Another way to append values to an existing list is by using the <code>insert()</code> function. For example <code>mylist.insert(index, new_element)</code>

```python
rows = int(input())
for i in range(0, rows):
    for ii in range (0, rows - i - 1):
        print(end = " ")
    for ii in range(0, i + 1):
        print("*", end = " ")
    print()
```
